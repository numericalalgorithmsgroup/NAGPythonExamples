# Question 3

Using a NAG routine produce two vectors of length _n_ both containing normally distributed numbers with mean 0 and standard deviation 2.  Plot the pairs of vectors against each other using the Python `matplotlib` library.

Now generate vectors of uniformly distributed numbers, again using a NAG routine, and plot them on the same axis.

```python
from naginterfaces.library import rand
from matplotlib import pyplot as plt

statecomm = rand.init_repeat(genid=2, seed=[32958])

n = 1000
X1 = rand.dist_normal(n, 0.0, 2.0 ** 2, statecomm)
Y1 = rand.dist_normal(n, 0.0, 2.0 ** 2, statecomm)

X2 = rand.dist_uniform(n, -4.0, 4.0, statecomm)
Y2 = rand.dist_uniform(n, -4.0, 4.0, statecomm)

plt.rcParams.update({'font.size': 14})
plt.scatter(X1, Y1, s=4, label='samples from normal')
plt.scatter(X2, Y2, s=4, label='samples from uniform')
plt.legend(loc='upper left')
fig = plt.gcf()
fig.set_size_inches(7.0, 7.0)

plt.show()
```
